7.Wkts:total wicket taken
8.BBI:Best bowling in innings(9/51     9 wicket and 51 run)<br>
9.BBM:Best bowling in matches(16/220   16 wicket and 220 run)<br>
10.Ave:Avarage run per wicket<br>
$\frac{run}{wkts}$<br>
11.Econ:Avarage run per over<br>
over=$\frac{Balls}{6}$<br>

# Renaming Column Name

```python
df = df.rename(columns = {'Player':'Player_name',
                          'Span':'Career_duration',
                          'Mat':'No_match_played', 
                          'Inns':'No_innings', 
                          'Balls':'No_balls_bowled',
                          'Runs': 'Runs_considered', 
                          'Wkts':'Total_wickets_taken', 
                          'BBI': 'Best_Bowling_in_Innings', 
                          'BBM':'Best_Bowling_in_Match',
                          'Ave':'Average_runs_per_wicket',
                          'Econ':'average_runs_per_over', 
                          'SR':'Strike_rate',
                          5: 'No_times_5_wickets_taken', 
                          10:'No_times_10_wickets_taken'})
display(df.head())
```
